```python
import pandas as pd
import matplotlib.pyplot as plt  
import numpy as np
```

```python
data = pd.read_csv("company_sales_data.csv")
data.head()
```

# Excercise - 1

```python
x = np.array(data["month_number"])
y = np.array(data["total_profit"])
plt.plot(x,y)
```

# Exercise - 2

```python
plt.plot(x , y, color="r", marker = "o", linewidth=3, linestyle='--')
plt.xlabel("Month Number")
plt.ylabel("Profit")
```

# Exercise - 3

```python
x1 = np.array(data["facecream"])
x2 = np.array(data["facewash"])
x3 = np.array(data["toothpaste"])
x4 = np.array(data["bathingsoap"])
x5 = np.array(data["shampoo"])
x5 = np.array(data["moisturizer"])

y = np.array(data["month_number"])

plt.plot(y, x1)
plt.plot(y, x2)
plt.plot(y, x3)
plt.plot(y, x4)
plt.plot(y, x5)
```

# Exercise - 4

# Exercise - 5

```python
y1 = np.array(data["facewash"])
y2 = np.array(data["facecream"])

x = np.array(data["month_number"])

plt.bar([a-0.25 for a in x],y1, align='edge', width= 0.25)
plt.bar([a+0.25 for a in x],y2, align='edge', width= 0.25)
```

# Exercise - 7

```python
x = np.array(data["total_profit"])
plt.hist(x)
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
```

# Exercise - 8

```python
p = [data["facecream"].sum(), data["facewash"].sum(), data["toothpaste"].sum(), data["bathingsoap"].sum(), 
data["shampoo"].sum(), data["moisturizer"].sum()]

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(p, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

# Exercise - 9

```python
f, axarr = plt.subplots(2, sharex=True)

y1 = np.array(data["facewash"])
y2 = np.array(data["bathingsoap"])
x = np.array(data["month_number"])

axarr[0].plot(x, y2, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(x, y1, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```
